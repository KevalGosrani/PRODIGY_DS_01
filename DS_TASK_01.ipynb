```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random population data (replace with your actual data)
np.random.seed(0)  # For reproducibility
population_data = np.random.normal(50, 10, 1000)  # Mean = 50, Standard Deviation = 10

# Create the histogram
plt.hist(population_data, bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('Population Values')
plt.ylabel('Frequency')
plt.title('Population Distribution Histogram')

# Display the histogram
plt.show()
```
